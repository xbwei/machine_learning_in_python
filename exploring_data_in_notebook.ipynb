## Importing the data

## A quick look of the data

pandas.DataFrame has  powerful statistics and visualizing functions. We can use the [pandas.DataFrame.describe](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.describe.html) to generate quick descriptive statistics of our data.

```python
df.describe()
```

We can also create [histograms](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.hist.html) to see the distributions of our data.

```python
df.hist(figsize=(20,20))
```

We can also create [scatter plots](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.scatter.html) to see the relationships among all columns. 

```python
from  pandas.plotting import scatter_matrix
scatter_matrix(df,figsize=(20, 20), diagonal='kde')
```

## Exploring the data

We want to see how many different types of houses are in our records:

```python
df.groupby('house_type').count()
```

Let's compare the average prices and views for each type of houses.

```python
df.groupby('house_type').mean()
```

We can also visualize the difference by creating histograms for each type of houses.

```python
df.groupby('house_type').hist(figsize=(12,4),column=['price','views'])
```

We want to see how the ages of hosues determine the prices. Here we use a scatterplot. 

```python
df.plot.scatter(x='built_in',y='price')
```

Since house prices are largely determined by sizes, ages and other factors, let's see how the house ages determine the house *unite price*, i.e., price/size. 
Let's calculate the unit price first, and then add the results to a new column called *unit_price*

```python
df['unit_price']=df['price']/df['area']
```

For each year, we calculate the average unite price, and assign the result to a new dataframe

```python
mean_price =df.groupby('built_in').mean()
```

[Line](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.line.html) chart is a good choice for visualizing trends. Let's use a line chart to see how the unite prices is determined by the house ages.

```python
mean_price.plot.line(y='unit_price')
```
